```python
import numpy as np
from scipy.special import expit, logit
import matplotlib.pyplot as plt
```

```python
# Define the sigmoid function
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

# Define the inverse sigmoid (logit) function
def inverse_sigmoid(y):
    return np.log(y / (1 - y))

# Generate test data
x = np.linspace(-10, 10, 100)
y = sigmoid(x)  # Apply sigmoid
x_reconstructed = inverse_sigmoid(y)  # Apply inverse sigmoid

# Plot results
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.plot(x, y, label="Sigmoid Function")
plt.xlabel("Input")
plt.ylabel("Sigmoid Output")
plt.legend()
plt.grid()

plt.subplot(1, 2, 2)
plt.plot(y, x_reconstructed, label="Inverse Sigmoid (Logit)", color="red")
plt.xlabel("Sigmoid Output")
plt.ylabel("Reconstructed Input")
plt.legend()
plt.grid()

plt.suptitle("Testing Sigmoid and Inverse Sigmoid Functions")
plt.show()
```
